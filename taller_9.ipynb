```python
def outlier(dfc,columnas_numericas):#metodo para graficar tablas bigotes
  fig,axes=plt.subplots(nrows=3,ncols=4,figsize=(10,10))
  axes=axes.flat

  for i, columna in enumerate(columnas_numericas):

    dfc[columna].plot.box(ax=axes[i])
    axes[i].set_title(columna, fontsize=10,fontweight="bold")
    axes[i].tick_params(labelsize=6)

  fig.tight_layout()
  fig.delaxes(axes[11])
  return
outlier(dfc,columnas_numericas)


```

```python
for i, columna in enumerate(columnas_numericas):
  cuantiles_mt=dfc[columna].quantile([0.25,0.5,0.75])
  Q1=cuantiles_mt[0.25]
  Q3=cuantiles_mt[0.75]
  IQR=Q3-Q1
  bigotesuper=Q3+(IQR*1.5)
  bigoteinfer=Q1-(IQR*1.5)
  print(columna,bigotesuper,bigoteinfer)
```

```python
#filtrar el dataframe antiguedad (outlayer)
def bigotes(dfc,columna):
  cuantiles_mt = dfc[columna].quantile([0.25,0.5,0.75])
  cuantiles_mt
  Q1 = cuantiles_mt[0.25]
  Q3 = cuantiles_mt[0.75]

  #IQR es rango intercuartil

  IQR = Q3-Q1
  bigotesuper = Q3+(IQR*1.5)
  bigoteinfer = Q1-(IQR*1.5)
  return bigotesuper,bigoteinfer
```

```python
#Quitar oulaiers de antiguedad

bigotesuper,bigoteinfer = bigotes(dfc,"antiguedad")
print(bigoteinfer,bigotesuper)
#dfc[dfc.antiguedad>bigotesuper]
```

```python
from sklearn.feature_selection import SelectKBest, f_regression

k=12
seleccionador=SelectKBest(f_regression,k=k)
X_trainB=seleccionador.fit_transform(X_train,y_train)
print(X_train.columns,len(X_train.columns))
x_test_columns=X_train.columns[seleccionador.get_support()]
print(X_train.columns[seleccionador.get_support()])
```

```python
from sklearn.linear_model import LinearRegression

regresion=LinearRegression()
modelo_regresion=regresion.fit(X_trainB,y_train)
jb.dump(modelo_regresion,"miprimermillon.bin", compress=True)
```

```python
prediccion_Train=modelo_regresion.predict(X_trainB)
prediccion_Train=pd.DataFrame({'precio real':y_train,'prediccion':prediccion_Train})
prediccion_Train
```

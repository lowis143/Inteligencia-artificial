```python
#graficos de histograma de todas las variables cualitativas

columnas_categoricas=dfc.select_dtypes(include='object').columns
columnas_categoricas

#definir los parametros de la grafica
fig,axes=plt.subplots(nrows=3,ncols=3,figsize=(9,5))
axes=axes.flat

#recorrer todas las columnas
for i, columna in enumerate(columnas_categoricas):
  dfc[columna].value_counts().plot.barh(ax=axes[i])
  axes[i].set_title(columna, fontsize=7,fontweight="bold")
  axes[i].tick_params(labelsize=6)
  axes[i].set_xlabel("")

#delaxes borra los subplots que se deseen
fig.delaxes(axes[7])
fig.delaxes(axes[8])

#ajustar los tamaños automaticamente para que los subplots no queden montados
fig.tight_layout()

#Ajustar titulo a la grafica
fig.subplots_adjust(top=0.9)
#poner titulos a las graficas
fig.suptitle("Distribuición de las variables cualitativas",fontsize=10,fontweight="bold")
```
